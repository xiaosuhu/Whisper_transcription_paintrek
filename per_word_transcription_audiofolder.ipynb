```python
# Install faster-whisper
!pip install -q faster-whisper

# Import required libraries
import os
import pandas as pd
from faster_whisper import WhisperModel
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the folder path containing audio files
folder_path = "/content/drive/My Drive/WhisperVideo/AudioFiles"

# Load Faster Whisper Model (Use "cuda" if you have GPU, otherwise "cpu")
model = WhisperModel("large-v2", device="cuda")  # Change to "cpu" if needed

# List all audio files in the folder (Modify extensions if needed)
audio_files = [f for f in os.listdir(folder_path) if f.endswith((".mp3", ".wav", ".m4a"))]

# Check if any audio files were found
if not audio_files:
    raise FileNotFoundError(f"No audio files found in {folder_path}!")

# List to store all word timestamps
all_word_data = []

# Process each audio file
for audio_file in audio_files:
    audio_path = os.path.join(folder_path, audio_file)
    print(f"Processing: {audio_file}")

    # Transcribe with word timestamps
    segments, _ = model.transcribe(audio_path, word_timestamps=True)

    # Extract word timestamps
    for segment in segments:
        for word in segment.words:
            all_word_data.append([audio_file, word.word, word.start, word.end])

# Convert to DataFrame
df = pd.DataFrame(all_word_data, columns=["Filename", "Word", "Start Time (s)", "End Time (s)"])

# Define output CSV file path
csv_output_path = os.path.join(folder_path, "all_audio_timestamps.csv")

# Save all results to a CSV file
df.to_csv(csv_output_path, index=False)

print(f"CSV file saved: {csv_output_path}")
```
