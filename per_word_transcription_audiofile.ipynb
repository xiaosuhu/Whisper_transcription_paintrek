```python
# Install faster-whisper
!pip install -q faster-whisper

# Import required libraries
import os
import pandas as pd
from faster_whisper import WhisperModel
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the correct file path
audio_path = "/content/drive/My Drive/WhisperVideo/AudioFiles/skates.gest.wav"

# Check if the file exists
if not os.path.exists(audio_path):
    raise FileNotFoundError(f"File not found: {audio_path}")

# Load Faster Whisper Model (Use "cuda" if you have GPU, otherwise "cpu")
model = WhisperModel("large-v2", device="cuda")  # Change to "cpu" if needed

# Transcribe with word timestamps
segments, _ = model.transcribe(audio_path, word_timestamps=True)

# Store word timestamps in a list
word_data = []
for segment in segments:
    for word in segment.words:
        word_data.append([word.word, word.start, word.end])

# Convert to DataFrame
df = pd.DataFrame(word_data, columns=["Word", "Start Time (s)", "End Time (s)"])

# Define output CSV file path
csv_output_path = "/content/drive/My Drive/WhisperVideo/AudioFiles/skates_timestamps.csv"

# Save to CSV
df.to_csv(csv_output_path, index=False)

print(f"CSV file saved: {csv_output_path}")
```
